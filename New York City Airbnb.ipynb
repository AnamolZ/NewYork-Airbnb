```python
cols_to_keep = ['id', 'host_id', 'latitude', 'longitude', 'room_type', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count']
cleanData = df[cols_to_keep]
cols_to_drop = ["neighbourhood", "availability_365", "name", "host_name", "neighbourhood_group", "last_review"]
cleanData = df.drop(cols_to_drop, axis='columns')
```

```python
for i, row in cleanData.iterrows():
    if row.isna().any():
        cleanData.drop(i, inplace=True)
```

```python
columns_to_check = ['id', 'host_id', 'latitude', 'longitude', 'room_type', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count']
cleanData.drop_duplicates(subset=columns_to_check, keep='first', inplace=True)
```

```python
lat = []
long = []
for i, row in cleanData.iterrows():
    if 40.67 <= row['latitude'] <= 40.87 and -74.02 <= row['longitude'] <= -73.88:
        lat.append(row['latitude'])
        long.append(row['longitude'])

plt.scatter(lat, long, color='red', s=1)
plt.show()
```
